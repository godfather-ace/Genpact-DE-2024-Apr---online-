### Importing Libraries

```python
import requests
from bs4 import BeautifulSoup
import json
```

### Accessing the video using Youtube URL

```python
youtube_url = "https://www.youtube.com/watch?v=yWVNPPHoyNY"
response = requests.request("GET", youtube_url)
```

### Creating a Soup

```python
soup = BeautifulSoup(response.text, "html.parser")
body = soup.find_all("body")[0]
scripts = body.find_all("script")
scripts
```

### Identifying the scripts tag dictionary keys

```python
result = json.loads(scripts[0].string[30:-1])
result.keys()
```

### Extracting streamingData Keys

```python
result['streamingData'].keys()
```

### Extracting streamingData Formats

```python
result['streamingData']['formats']
```
