### Loading the necessary libraries

```python
from urllib.request import urlopen #used to open urls
from bs4 import BeautifulSoup #used for extracting data from html websites
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Accessing HTML content from a webpage

```python
url = "https://en.wikipedia.org/wiki/Jupiter"
html = urlopen(url)
```

### Parsing the HTML content of the webpage

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
type(soup)
```

### Getting the title of the webpage

```python
print(soup.title)
```

### Viewing the data in pretty format

```python
print(soup.prettify())
```

### find() and find_all() functions - BeautifulSoup 

```python
print(soup.find('h1'))
```

```python
print(soup.find_all('li', limit = 5))
```

```python
print(soup.find('h1').prettify())
```

### Printing all the table rows and table headers

```python
rows = soup.find_all('tr')
rows
```

```python
rows = soup.find_all('th')
rows
```

### Printing all the hyperlinks on webpage

```python
links = soup.find_all('a', href = True)
for i in links:
    print(i['href'])
```

```python
total_links = len(links)
print("Total number of weblinks on the webpage", total_links)
```

### Extracting text from DIV tags

```python
contents = soup.find_all('div')
for i in contents:
    print(i.text)
```

### Additional Operations

#### Printing the title string only

```python
print(soup.title.string)
```

#### Printing paragraph content

```python
for paragraph in soup.find_all('p'):
    print(str(paragraph.text))
```

#### Getting all the links of the webpage

```python
for url in soup.find_all('a'):
    print(url.get('href'))
```

```python
table_rows = soup.find_all('tr')
for tr in table_rows: 
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

#### Scrapping webpage data into a dataframe - Pandas (read_html function)

```python
dfs = pd.read_html("https://en.wikipedia.org/wiki/Jupiter", header= 0)
for df in dfs: 
    print(df)
```
